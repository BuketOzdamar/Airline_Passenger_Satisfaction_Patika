## Aykƒ±rƒ± Deƒüer Analizi

```python
numerical_cols = ["Age", "Flight Distance", "Departure Delay in Minutes", "Arrival Delay in Minutes"] # Hizmetleri katmadƒ±m √ß√ºnk√º onlar sadece 0-5 arasƒ±nda puanlandƒ±

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"\nüîç {col}")
    print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
    print(f"Alt Sƒ±nƒ±r: {lower_bound:.2f}, √úst Sƒ±nƒ±r: {upper_bound:.2f}")
    print(f"Aykƒ±rƒ± deƒüer sayƒ±sƒ±: {outliers.shape[0]}")
```

Sayƒ±sal deƒüi≈ükenler √ºzerinde yapƒ±lan IQR (Interquartile Range) temelli aykƒ±rƒ± deƒüer analizi sonucunda:  
Age (Ya≈ü) deƒüi≈ükeninde aykƒ±rƒ± deƒüer tespit edilmemi≈ütir. Bu, ya≈ü verisinin istatistiksel olarak dengeli daƒüƒ±ldƒ±ƒüƒ±nƒ± g√∂stermektedir.  
Flight Distance, Departure Delay ve Arrival Delay deƒüi≈ükenlerinde ise y√ºksek sayƒ±da aykƒ±rƒ± deƒüer g√∂zlenmi≈ütir:  
-Flight Distance: 2.287 aykƒ±rƒ± deƒüer  
-Departure Delay: 14.428 aykƒ±rƒ± deƒüer  
-Arrival Delay: 13.954 aykƒ±rƒ± deƒüer  
Bu aykƒ±rƒ± deƒüerlerin, √∂zellikle gecikme s√ºreleri ve u√ßu≈ü mesafeleri gibi u√ßucu ko≈üullardan etkilenebilecek √∂zelliklerde yoƒüunla≈ümasƒ± dikkat √ßekicidir.
√ñzellikle hava ko≈üullarƒ±, teknik arƒ±zalar veya nadiren ger√ßekle≈üen uzun mesafe u√ßu≈ülar bu ekstrem deƒüerlerin sebebi olabilir.

```python
# U√ßu≈ü Mesafesinde G√∂re Memnuniyet Oranƒ± 
# Memnun olanlar 1 olmayanlar 0 olsun. Buna g√∂re gruplayƒ±p u√ßu≈ü mesafesine g√∂re memuniyet oranlarƒ±nƒ± kar≈üƒ±la≈ütƒ±rƒ±yoruz.


df_copy['satisfaction_binary'] = df_copy['satisfaction'].map({
    'satisfied': 1,
    'neutral or dissatisfied': 0
})

bins = [0, 1000, 2000, 4000]
labels = ['Kƒ±sa (0‚Äì1000)', 'Orta (1001‚Äì2000)', 'Uzun (2001‚Äì4000)']
df_copy['Mesafe Grubu'] = pd.cut(df_copy['Flight Distance'], bins=bins, labels=labels, include_lowest=True)

memnuniyet_sayilari = df_copy.groupby(['Mesafe Grubu', 'satisfaction_binary']).size().unstack(fill_value=0)
memnuniyet_sayilari.columns = ['Memnun Deƒüil', 'Memnun']
print(memnuniyet_sayilari)
```
