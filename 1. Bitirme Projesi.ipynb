Alternatif olarak, eksik değerlerin ortalama, medyan ya da standart sapma gibi istatistiksel yöntemlerle doldurulması mümkün olsa da; mevcut durumda veri kaybı minimal olduğu için doğrudan çıkarım tercih edilmiştir.

Ayrıca, eksik verilerin neden oluştuğu da değerlendirilmelidir. Bu eksiklikler, veri giriş hatalarından mı, yoksa sistematik bir durumdan mı kaynaklanıyor incelenmelidir.

## Aykırı Değer Analizi

```python
numerical_cols = ["Age", "Flight Distance", "Departure Delay in Minutes", "Arrival Delay in Minutes"] # Hizmetleri katmadım çünkü onlar sadece 0-5 arasında puanlandı

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"\n🔍 {col}")
    print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
    print(f"Alt Sınır: {lower_bound:.2f}, Üst Sınır: {upper_bound:.2f}")
    print(f"Aykırı değer sayısı: {outliers.shape[0]}")
```

Sayısal değişkenler üzerinde yapılan IQR (Interquartile Range) temelli aykırı değer analizi sonucunda:  
Age (Yaş) değişkeninde aykırı değer tespit edilmemiştir. Bu, yaş verisinin istatistiksel olarak dengeli dağıldığını göstermektedir.  
Flight Distance, Departure Delay ve Arrival Delay değişkenlerinde ise yüksek sayıda aykırı değer gözlenmiştir:  
-Flight Distance: 2.287 aykırı değer  
-Departure Delay: 14.428 aykırı değer  
-Arrival Delay: 13.954 aykırı değer  
Bu aykırı değerlerin, özellikle gecikme süreleri ve uçuş mesafeleri gibi uçucu koşullardan etkilenebilecek özelliklerde yoğunlaşması dikkat çekicidir.
Özellikle hava koşulları, teknik arızalar veya nadiren gerçekleşen uzun mesafe uçuşlar bu ekstrem değerlerin sebebi olabilir.

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in numerical_cols:
    plt.figure(figsize=(8, 1.5))
    sns.boxplot(x=df[col], color="coral")
    plt.title(f"{col} için Boxplot")
    plt.show()
```

Uçuş mesafesinde 4000 km civarı gözlemler aykırı gibi dursa da, veri setinde uluslararası uçuşlar varsa bu değerler gerçektir. Bu nedenle bağlamsal olarak incelenmeli, filtrelemeden önce uçuş tipi bilinmeli.   
Kalkış ve varış gecikmeleri incelendiğinde 1500 dakikayı aşan aykırı değerler görülmekte. Bu değerler normal uçuş süresinden çok yüksek olsa da, hava durumu, iptaller veya operasyonel problemler gibi faktörlerle açıklanabilir. Bu yüzden aykırı değerlerin veri hatası olup olmadığı, ekstra bilgi ile birlikte değerlendirilmelidir.

```python
df_copy = df
```

```python
# Uçuş Mesafesinde Göre Memnuniyet Oranı 
# Memnun olanlar 1 olmayanlar 0 olsun. Buna göre gruplayıp uçuş mesafesine göre memuniyet oranlarını karşılaştırıyoruz.


df_copy['satisfaction_binary'] = df_copy['satisfaction'].map({
    'satisfied': 1,
    'neutral or dissatisfied': 0
})

bins = [0, 1000, 2000, 4000]
labels = ['Kısa (0–1000)', 'Orta (1001–2000)', 'Uzun (2001–4000)']
df_copy['Mesafe Grubu'] = pd.cut(df_copy['Flight Distance'], bins=bins, labels=labels, include_lowest=True)

memnuniyet_sayilari = df_copy.groupby(['Mesafe Grubu', 'satisfaction_binary']).size().unstack(fill_value=0)
memnuniyet_sayilari.columns = ['Memnun Değil', 'Memnun']
print(memnuniyet_sayilari)
```

```python
memnuniyet_sayilari.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Uçuş Mesafelerine Göre Memnuniyet Dağılımı')
plt.ylabel('Yolcu Sayısı')
plt.xlabel('Uçuş Mesafesi')
plt.legend(title='Memnuniyet Durumu')
plt.tight_layout()
plt.show()
```
